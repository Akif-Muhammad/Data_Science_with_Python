```python
# imort libraries
import seaborn as sns 
import matplotlib.pyplot as plt
```

## Adding title

```python
# Load Dataset
phool=sns.load_dataset("iris")
phool
# draw a line plot(line plot 2 numarical variables k drmyan bnta hai)
sns.lineplot(x="sepal_length", y="sepal_width", data=phool)
plt.title("line plot on iris dataset")
plt.show()
```

## Adding limit

```python
phool=sns.load_dataset("iris")
phool
# draw a line plot(line plot 2 numarical variables k drmyan bnta hai)
sns.lineplot(x="sepal_length", y="sepal_width", data=phool)
plt.title("line plot on iris dataset")
plt.xlim(2)
plt.ylim(1)
plt.show()
```

## Size of figure

```python
# imort libraries
import seaborn as sns 
import matplotlib.pyplot as plt
# load data set
phool=sns.load_dataset("iris")
# draw a line plot
sns.lineplot(x="sepal_length", y="sepal_width", data=phool)
# add title
plt.title("line plot on iris dataset")
# add figure
plt.figure(figsize=(2,1))
plt.show()
```

# _**Barplot**_

```python
# imort libraries
import seaborn as sns 
import matplotlib.pyplot as plt
# load data set
phool=sns.load_dataset("iris")
# draw a bar plot
sns.barplot(x="species", y="petal_width", data=phool)
# add title
plt.title("Bar plot on iris dataset")
plt.show()
phool
```

```python
import seaborn as sns 
import matplotlib.pyplot as plt
# load data set
kashti=sns.load_dataset("titanic")
# draw a bar plot
sns.barplot(x="sex", y="alone",hue="who", data=kashti)
# add title
plt.title("Bar plot on titanic dataset")
plt.show()
kashti
```

```python
import seaborn as sns 
import matplotlib.pyplot as plt
# load data set
kashti=sns.load_dataset("titanic")
# draw a bar plot| in order shape and we can add colors
sns.barplot(x="sex", y="alone",hue="who", data=kashti,order=["female", "male"],color= "green")
# add title
plt.title("Bar plot on titanic dataset")
plt.show()
```

```python
import seaborn as sns 
import matplotlib.pyplot as plt
# load data set
kashti=sns.load_dataset("titanic")
# remove error bar and palettes color
sns.barplot(x="sex", y="alone",hue="who", data=kashti,order=["female", "male"],color= "green", ci=None,
palette='bright')
# add title
plt.title("Bar plot on titanic dataset")
plt.show()
```

```python
import seaborn as sns 
from numpy import median
import matplotlib.pyplot as plt
# load data set
kashti=sns.load_dataset("titanic")
# estimate (need to numarical values for estimates) and saturation(its mean that intensity of color)
sns.barplot(x="class", y="fare",hue="sex", data=kashti, estimator=median,saturation=0.2)
# add title
plt.title("Bar plot on titanic dataset")
plt.show()
```

```python
# horizontal plot
# (make a horizontal plot simpaly numaric parameters adjust on x axis and
#  categorical parameters adjust on y axis)
import seaborn as sns
from numpy import median
import matplotlib.pyplot as plt
# load data set
kashti=sns.load_dataset("titanic")
# estimate (need to numarical values for estimates) and saturation(its mean that intensity of color)
sns.barplot(x="fare", y="class",hue="sex", data=kashti, estimator=median,saturation=3)
# add title
plt.title("Bar plot on titanic dataset")
plt.show()
```
