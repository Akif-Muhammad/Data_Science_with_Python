#  Boxplot for Data Visualization

```python
import seaborn as sns
import pandas as pandas
import numpy as numpy
import matplotlib.pyplot as plt
kashti=sns.load_dataset('titanic')
kashti.head()
```

```python
sns.boxplot(x='survived',
y='age', data=kashti)
```

```python
p1=sns.boxplot(x='survived',
y='age',showmeans='True', data=kashti)
```

```python
# you can make means symbol your own choice:
p1=sns.boxplot(x='survived',
y='age',showmeans='True',
meanprops={"marker":"*",
"markersize":"10",
"markeredgecolor":"red"},
 data=kashti)
#  show Labels
plt.xlabel("How many survived", size=10,weight="bold" , color= "green"),
plt.ylabel("Age of the passengers", size=12,weight="bold" , color= "blue"),
plt.title("[Boxplot On Titanic Dataset]", size= 20, weight="bold" , color= "red")

plt.show()
```

# Assinment : what is facet wrap and facet grid ?





```python
import seaborn as sns
import matplotlib.pyplot as plt

kashti=sns.load_dataset('titanic')
#  main section of assignment
plot=sns.FacetGrid( col="sex", hue="fare", data=kashti)
plot.map(plt.scatter, "survived", "age", alpha=.7, edgecolor="r").add_legend()
plt.show()
```
