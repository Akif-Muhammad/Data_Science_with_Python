```python
# 1st step import libraries
# involved in data viualization
import seaborn as sns 
import matplotlib.pyplot as plt
# 2nd step set a theme 
sns.set_theme(style="ticks", color_codes=True)

# 3rd import dataset you can also import your own data
kashti= sns.load_dataset("titanic")
kashti
# kashti.info
# kashti.shape
# kashti.describe
# 4th step plot basic graph(count plot)
# countplot ki left side mein already cout aa jata hai we will not put any y exis here .
k=sns.countplot(x="sex", data=kashti)
plt.show()

# 5th step plot graph with two variables (count plot)
k=sns.countplot(x="sex", data=kashti,hue="class")
plt.show()

# 6th step basic plot graph with 2 variables (count plot)with titles
k=sns.countplot(x="sex", data=kashti,hue="class")
k.set_title("plot for basic counting")
plt.show()
```
